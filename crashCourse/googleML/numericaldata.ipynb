```python
import pandas as pd
```

```python
# The following lines adjust the granularity of reporting.
pd.options.display.max_rows = 10
pd.options.display.float_format = "{:.1f}".format
```

```python
training_df = pd.read_csv(filepath_or_buffer="https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv")
```

```python
training_df.describe()
```

```python
training_df.hist(figsize=(12, 10), bins=50)
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each column
Q1 = training_df.quantile(0.25)
Q3 = training_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers: values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers = ((training_df < (Q1 - 1.5 * IQR)) | (training_df > (Q3 + 1.5 * IQR)))

# Display the count of outliers for each column
outliers.sum()
```
